```python
fuel_economy.sample(30)
```

```python
vehicle_sample_set=[fuel_economy.sample(30) for i in range(1,10)]
vehicle_sample_set[1]
```

```python
vehicle_sample_set[1].mean()
```

```python
vehicle_sample_set[2].mean()
```

```python
for sample in vehicle_sample_set:
    print(sample.mean())
```

```python
means = [sample['Combined_MPG'].mean() for sample in vehicle_sample_set]
```

```python
standard_errors=[sem(sample['Combined_MPG']) for sample in vehicle_sample_set]
standard_errors
```

```python
x_axis = np.arange(0, len(vehicle_sample_set), 1) + 1
```

```python
# Setting up the plot
fig, ax = plt.subplots()
ax.errorbar(x_axis, means, standard_errors, fmt="o")
ax.set_xlim(0, len(vehicle_sample_set) + 1)
ax.set_ylim(20,28)
ax.set_xlabel("Sample Number")
ax.set_ylabel("Mean MPG")
plt.show()
```
