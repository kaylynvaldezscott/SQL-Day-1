```python
temperatures.quantile([.25,.5,.75])
```

```python
temperatures.quantile([.25,.5,.75])[0.25]
```

```python
loweq=temperatures.quantile([.25,.5,.75])[0.25]
upperq=temperatures.quantile([.25,.5,.75])[0.75]
```

```python
iqr= upperq-loweq
```

```python
lower_bound=  loweq-1.5*iqr
upper_bound=  upperq+1.5*iqr
```

```python
loweq,upperq,iqr,  lower_bound,upper_bound
```
