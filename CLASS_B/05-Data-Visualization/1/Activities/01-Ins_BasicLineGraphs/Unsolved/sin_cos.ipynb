```python
# Create our x_axis numpy array
x_axis=np.arange(0,5,0.1)
x_axis
```

```python
# Creates a numpy array based on the sin of our x_axis values
sin=np.sin(x_axis)
sin
```

```python
# Creates a numpy array based on the cos of our x_axis values
cos=np.cos(x_axis)
```

```python
plt.plot(x_axis,cos)
plt.xlabel("X")
```

```python
plt.plot(x_axis,sin)
plt.plot(x_axis,cos)
plt.xlabel("X")
plt.ylabel("Y")
```
