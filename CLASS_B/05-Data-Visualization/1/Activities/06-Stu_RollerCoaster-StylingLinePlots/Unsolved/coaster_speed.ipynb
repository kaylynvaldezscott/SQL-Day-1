```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
time = np.arange(0,130,10)
danger_drop_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# Plot the charts and apply some styling
plt.plot(time,danger_drop_speeds,color='red',marker='o',label='danger')
plt.plot(time,railgun_speeds,color='blue',marker='s',label='railgun')
```

```python
plt.xlabel("Months")
plt.ylabel("Y")
```

```python
plt.legend()
```
